Neural Network

```python
from keras.models import Model, Sequential
from keras.layers import Dense, Flatten, Conv2D, MaxPooling2D
from keras.layers.normalization import BatchNormalization
```

```python
from keras.utils import to_categorical
y_train_one_hot = to_categorical(y_train)
y_test_one_hot = to_categorical(y_test)
```

```python
activation = "sigmoid"
SIZE = 128
classifier = Sequential()
classifier.add(Conv2D(32, 3, activation = "relu", padding = "same", input_shape = (SIZE, SIZE, 3)))
classifier.add(BatchNormalization())

classifier.add(Conv2D(32, 3, activation = "relu", padding = 'same', kernel_initializer = 'he_uniform'))
classifier.add(BatchNormalization())
classifier.add(MaxPooling2D())

classifier.add(Conv2D(64, 3, activation = "relu", padding = 'same', kernel_initializer = 'he_uniform'))
classifier.add(BatchNormalization())

classifier.add(Conv2D(64, 3, activation = "relu", padding = 'same', kernel_initializer = 'he_uniform'))
classifier.add(BatchNormalization())
classifier.add(MaxPooling2D())

classifier.add(Flatten())
```

```python
x = classifier.output  
x = Dense(128, activation = "relu", kernel_initializer = 'he_uniform')(x)
prediction_layer = Dense(2, activation = 'softmax')(x)
```

```python
classifier = Model(inputs=classifier.input, outputs=prediction_layer)
classifier.compile(optimizer='rmsprop',loss = 'categorical_crossentropy', metrics = ['accuracy'])
print(classifier.summary())
```

```python
classifier.fit(x_train, y_train_one_hot, epochs=5, validation_data = (x_test, y_test_one_hot))
```
